```python
from rembg import remove
from PIL import Image
import os
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python

os.makedirs('test1', exist_ok=True)
os.makedirs('test2', exist_ok=True)
```

```python
# Replace this URL with a direct link to the image file
img_path = '/content/test.PNG'
```

```python
if os.path.exists(img_path):
    img = Image.open(img_path)

    # Display the original image
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(img)
    plt.axis('off')

    # Check if the image has an alpha channel (transparency)
    if img.mode == 'RGBA':
        # Convert the image to RGB mode (removing alpha channel)
        img = img.convert('RGB')

    # Save the image as JPEG with the correct file extension
    img.save('test1/shivam.jpg', format='jpeg')

    # Display the image after background removal
    plt.subplot(1, 2, 2)
    plt.title('Image After Background Removal')

    # Read the saved JPEG image and pass it to rembg
    input_image = open('test1/shivam.jpg', 'rb').read()
    subject = remove(input_image)
    plt.imshow(Image.open(BytesIO(subject)))
    plt.axis('off')

    # Show the plots
    plt.show()
else:
    print(f"Image file not found: {img_path}")
```
